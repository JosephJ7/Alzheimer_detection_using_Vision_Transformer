```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
pic = load_img('C:/Users/sameer/Downloads/archive/Alzheimer_s Dataset/train/ModerateDemented/moderateDem1.jpg')
pic
```

```python
pic_array = img_to_array(pic)
```

```python
pic_array.shape
```

```python
pic_array = pic_array.reshape((1,) + pic_array.shape) # Converting into 4 dimension array
```

```python
pic_array.shape
```

```python
#generate images now
#batch size= At a time how many images should be created

count=0
for batch in datagen.flow(pic_array,batch_size=1,save_to_dir='C:/Users/sameer/Downloads/archive/Alzheimer_s Dataset/train/Moderate_NewAugumented',save_prefix='mod_dem',save_format='jpg'):
    count=count+1
    if count>10:
        break
```
